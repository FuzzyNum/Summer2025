```python
board = [["5","3",".",".","7",".",".",".","."],["6",".",".","1","9","5",".",".","."],[".","9","8",".",".",".",".","6","."],["8",".",".",".","6",".",".",".","3"],["4",".",".","8",".","3",".",".","1"],["7",".",".",".","2",".",".",".","6"],[".","6",".",".",".",".","2","8","."],[".",".",".","4","1","9",".",".","5"],[".",".",".",".","8",".",".","7","9"]]
expanded_list = []
for row in board:
    for i in row:
        expanded_list.append(i)

def block(i):
            column = i%9
            row = i//9
            return column//3 + 3*(row//3)
        
def already_in_row(row=0, num=1):
    for i in [expanded_list[j] for j in range(9*row,9*(row+1))]:
        if i==f'{num}':
            return True
    return False

def already_in_col(col=0, num=1):
    for i in [expanded_list[j] for j in range(col,len(expanded_list),9)]:
        if i==f'{num}':
            return True
    return False
```
