```python
board = [["5","3",".",".","7",".",".",".","."],["6",".",".","1","9","5",".",".","."],[".","9","8",".",".",".",".","6","."],["8",".",".",".","6",".",".",".","3"],["4",".",".","8",".","3",".",".","1"],["7",".",".",".","2",".",".",".","6"],[".","6",".",".",".",".","2","8","."],[".",".",".","4","1","9",".",".","5"],[".",".",".",".","8",".",".","7","9"]]
expanded_list = []

class Board:
    def __init__(self,board):
        self.board = board
        self.digits = ["1","2","3","4","5","6","7","8","9"]
        self.counts = {i:0 for i in self.digits}
        self.expanded_list = self.build_expanded_list()
        
    
    def build_expanded_list(self):
        expanded_list = []
        for row in self.board:
            for i in row:
                expanded_list.append(i)
                if i!=".":
                    self.counts[i] += 1
        return expanded_list
    
    def block(self,i):
        column = i%9
        row = i//9
        return column//3 + 3*(row//3)
    
    def row(self,i):
        return i//9

    def col(self,i):
        return i%9
    
    def find_empty(self,board_list):
        for i in range(81):
            if board_list[i] == ".":
                return i
        return None
    
    def solve_backtrack(self,board_list):
        pos = self.find_empty(board_list)
        
        if pos is None:
            return True

        for num in self.candidates(board_list,pos):
            board_list[pos] = num

            if self.solve_backtrack(board_list):
                board_list[pos] = "." 
        return False
    
    def solve(self):
        board_copy = self.expanded_list.copy()
        self.solve_backtrack(board_copy)
        return board_copy


    def already_in_row(self, board_list, row=0, num=1):
        for k in [board_list[j] for j in range(9*(row),9*((row)+1))]:
            if k==f'{num}':
                return True
        return False    
    
    def already_in_col(self, board_list, col=0, num=1):
        for k in [board_list[j] for j in range((col),len(board_list),9)]:
            if k==f'{num}':
                return True
        return False

    def already_in_block(self, board_list, block_num, num=1):
        block_row = block_num // 3
        block_col = block_num % 3

        for r in range(block_row * 3, block_row * 3 + 3):
            for c in range(block_col * 3, block_col * 3 + 3):
                if board_list[r * 9 + c] == str(num):
                    return True
        return False
    
    def candidates(self,board_list,i):
        return [j for j in self.digits if not self.already_in_row(board_list,self.row(i),num=j) and not self.already_in_col(board_list,self.col(i), num=j) and not self.already_in_block(board_list,self.block(i),num=j)]
    
    def solve_row(self, big_list, row):
        for i in range(0,9):
            if big_list[9*row+i]!=".":
                pass
            
            else:
                index = 0
                while len(self.candidates(big_list,9*row+i))!=0 and index<len(self.candidates(big_list,9*row+i)):
                    big_list[9*row+i]=self.candidates(big_list,9*row+i)[index]
                    self.solve_row(big_list,row)
                    index+=1
        return big_list[9*row:9*row+9]
            
            
    def solve_col(self,big_list,col):
        for i in range(0,9):
            if big_list[9*i+col]!=".":
                pass
            else:
                index = 0
                while len(self.candidates(big_list, 9*i+col))!=0 and index<len(self.candidates(big_list,9*i+col)):
                    big_list[9*i+col]=self.candidates(big_list,9*i+col)[index]
                    self.solve_row(big_list,col)
                    index+=1
        return [big_list[j] for j in range((col),len(big_list),9)]
            
        
    
    def solve_block(self,big_list,block):
        block_row = block // 3
        block_col = block % 3

        for r in range(block_row * 3, block_row * 3 + 3):
            for c in range(block_col * 3, block_col * 3 + 3):
                index = 0
                while len(self.candidates(big_list, r*9+c))!=0 and index<len(self.candidates(big_list,9*r+c)):
                    big_list[9*r+c] = self.candidates(big_list, 9*r+c)[index]
                    self.solve_block(big_list,block)
                    index+=1
        return big_list
                






        
    
```
