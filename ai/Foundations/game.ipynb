```python
import random
```

```python
sports_choice = {"lacrosse":50,"squash":50,"basketball":90,"baseball":80,"soccer":90,"football":90,"tennis":70}
school_choice = {"Harvard":90,"Stanford":90,"MIT":94,"Yale":85,"Oxford":96,"Berkeley":80,"UCLA":80, "UNC":60,"UVA":60}
```

```python
class Game:
    def __init__(self):
        self.country = random.choice(country_names)
        self.age = 0
        self.occupation = "child"
        self.dating = False
        self.school = ""
        self.wealth = random.choice([float(i) for i in range(0,100)])
        self.intelligence = random.choice([float(i) for i in range(0,100)])
        self.social = random.choice([float(i) for i in range(0,100)])
        self.athletic_talent = random.choice([float(i) for i in range(0,100)])
        self.sport = ""
        self.luck = random.choice([float(i) for i in range(0,100)]) + self.social/10
        self.children = []
        self.learned = []
        self.bought = []
        self.life = []
    def new_life(self):
        self.country = random.choice(country_names)
        self.age = 0
        self.school = ""
        self.dating = False
        self.occupation = "child"
        self.wealth = random.choice([float(i) for i in range(0,100)])
        self.intelligence = random.choice([float(i) for i in range(0,100)])
        self.social = random.choice([float(i) for i in range(0,100)])
        self.athletic_talent = random.choice([float(i) for i in range(0,100)])
        self.sport = ""
        self.luck = random.choice([float(i) for i in range(0,100)])
        self.children = []
        self.learned = []
        self.bought = []
        self.life = []
    def __repr__(self):
        return f'{self.age}-year old from {self.country} with wealth {self.wealth}. Current status: {self.occupation}. Has done these things: {self.life} \nIntelligence: {self.intelligence}\nWealth: {self.wealth}\nCountry: {self.country}\nLearned: {self.learned}\nBought: {self.bought}'
    def age_up(self):
        self.age+=1
    def get_actions(self):
        if self.age<18:
            return ["done","help","age","show","choose_sport","train_sport","learn"]
        elif self.age==18:
            return ["done","help","age","show","choose_sport","train_sport","learn","apply"]
        elif self.age<22:
            return ["done","help","age","show","learn"]
        elif self.age==22:
            return ["done","help","age","show","apply_grad","apply_job","buy","date","move"]
        elif self.age<80:
            return ["done","help","age","show","kid","date","promotion","apply_job","buy","free_time","move"]
        
    def play(self):
        while self.age<80:
            available = self.get_actions()
            action = input("Choice: ")
            if action in available:    
                if action=="done":
                    self.age_up()
                    if self.occupation!="student":
                        self.wealth*=1.01
                    print(self)
                    pass
                elif action=="help":
                    print(self.get_actions())
                elif action == "age":
                    print(self.age())
                elif action == "show":
                    print(self)
                elif action=="choose_sport":
                    self.sport = input("Choose sport: ")
                elif action=="train_sport":
                    self.athletic_talent*=1.01
                elif action=="learn":
                    self.intelligence*=1.01
                    self.luck*=1.01
                    self.learned.append(input("Learned what: "))
                elif action=="apply" or action=="apply_grad":
                    school = input("What School: ")
                    if school in school_choice.keys():
                        if self.intelligence>=school_choice.get(school):
                            print("You got in.")
                            self.school = school
                        else:
                            print("You got rejected.")
                    else:
                        print("You got in.")
                        self.school = school
                elif action=="buy":
                    thing = input("Buy what? ")
                    price = input("For how much? ")
                    if float(price)>(self.wealth*1000):
                        print("You can't afford this.")
                    else:
                        print("You just bought a "+thing)
                        self.bought.append(thing)
                        self.wealth -= 1*float(price)/self.wealth
                elif action == "date":
                    self.dating = True
                elif action == "kid":
                    if self.dating:
                        print("You now have a kid.")
                        child = Game()
                        child.wealth = self.wealth
                        child.country = self.country
                        child.athletic_talent = self.athletic_talent
                    else:
                        print("Kid with who?")
                elif action == "move":
                    country = input("To Where? ")
                    self.country = country
                    for kid in self.children:
                        kid.country = country
                elif action == "apply_job":
                    self.occupation = input("What job: ")
                elif action == "promotion":
                    title = input("To What: ")
                    yesorno = random.choice([True, False])
                    if yesorno:
                        print("You have been promoted.")
                        self.occupation = title
                elif action == "free_time":
                    thing = input("Do what: ")
                    self.life.append(thing)
                elif action == "view_kids":
                    for i in self.children:
                        print(i)
        yn = input("You have died, would you like to become your child? ")
        if yn == "y":
            for i in self.children:
                print(i)
            kid = input("Which one (number)? ")
            self = self.children[kid]
        else:
            print("Thanks for playing.")

        
        


                



    

        
```
